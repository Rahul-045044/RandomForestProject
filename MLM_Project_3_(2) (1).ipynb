##Dataset Description:

Valence: Valence represents a measure of the musical positiveness conveyed by a track, ranging from 0 to 1. A value closer to 1 indicates a more positive emotional tone, while a value closer to 0 represents a less positive tone.

Year: Year denotes the release year of the track, providing temporal information about the music's historical context and evolution over time.

Acousticness: Acousticness quantifies the acoustic characteristics of the track on a scale from 0 to 1. A value of 0 suggests a predominantly electronic or synthesized sound, while a value of 1 indicates a purely acoustic recording.

Danceability: Danceability measures how suitable a track is for dancing, ranging from 0 to 1. Higher values imply tracks with rhythms and beats conducive to dancing.

Duration_ms: Duration_ms specifies the duration of the track in milliseconds, providing insight into the length of the musical composition.

Energy: Energy reflects the intensity and vigor of the track's musical elements, ranging from 0 to 1. Higher values indicate more energetic and lively compositions.

Instrumentalness: Instrumentalness quantifies the likelihood that the track contains no vocals, ranging from 0 to 1. A higher value suggests a higher probability of instrumental-only compositions.

Liveness: Liveness measures the presence of a live audience in the recording, with values ranging from 0 to 1. A higher value indicates a higher probability of a live performance.

Loudness: Loudness denotes the overall loudness of the track in decibels (dB), providing information about its volume and intensity.

Popularity: Popularity indicates the popularity of the track, reflecting its reception and engagement among listeners.

Speechiness: Speechiness measures the presence of spoken words in the track, ranging from 0 to 1. Higher values indicate tracks with more spoken-word elements, such as podcasts or spoken-word performances.

Tempo: Tempo represents the speed or pace of the track in beats per minute (BPM), providing information about its rhythmic structure and tempo.

Cluster number: cluster 0 - Versatile and Balanced, cluster 1- Energetic and Dynamic, cluster 2- Dynamic and Expressive.

1.1 If the acousticness feature is less than or equal to 0.48, further splitting occurs based on the speechiness_mn feature:

1.1.2 If speechiness_mn is greater than 0.92, the class is assigned as 2.

1.2. If the acousticness feature is greater than 0.48, further splitting occurs based on the speechiness_mn feature:

1.2.1 If speechiness_mn is less than or equal to 0.87, the class is assigned as 0 (likely representing the Versatile and Balanced cluster).

1.2.2 If speechiness_mn is greater than 0.87, the class is assigned as 2.

Right Branch (mode_y > 0.50):

2.1 If explicit_y is less than or equal to 0.50, further splitting occurs based on the duration_ms_mn feature.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as sps

cols = df_noncat.columns
# Initialize an empty DataFrame to store p-values
p_values_df = pd.DataFrame(index=cols, columns=cols)

# Iterate over each pair of columns
for col1 in cols:
    for col2 in cols:
        # Skip if the columns are the same or if any of them is non-numeric
        if col1 == col2 or not (df_noncat[col1].dtype == 'float64' and df_noncat[col2].dtype == 'float64'):
            continue

        # Drop rows with NaN values in both columns
        filtered_df = df_noncat[[col1, col2]].dropna()

        # Perform Pearson correlation test
        _, p_value = sps.pearsonr(filtered_df[col1], filtered_df[col2])

        # Store the p-value in the DataFrame
        p_values_df.loc[col1, col2] = p_value

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(p_values_df.astype(float), annot=True, cmap='coolwarm', fmt=".3f", cbar_kws={'label': 'p-value'})
plt.title('Heatmap of Pearson Correlation P-values')
plt.xlabel('Variable')
plt.ylabel('Variable')
plt.show()
```

```python
# Initialize StratifiedShuffleSplit with desired test size and random state
stratified_split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=45044)
# Perform the stratified split to get training and testing indices
for i, j in stratified_split.split(music_inputs, music_outputs):
    train_music_inputs, test_music_inputs = music_inputs.iloc[i], music_inputs.iloc[j]
    train_music_outputs, test_music_outputs = music_outputs.iloc[i], music_outputs.iloc[j]
```

```python
# Decision Tree : Prediction Evaluation (Testing Subset)
dtc_predict_conf_mat = pd.DataFrame(confusion_matrix(test_music_outputs, dtc_model_predict_test));print(dtc_predict_conf_mat)
dtc_predict_perf = classification_report(test_music_outputs, dtc_model_predict_test);  print(dtc_predict_perf)
```
